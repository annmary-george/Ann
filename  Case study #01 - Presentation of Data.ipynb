# Case Study #01 - Presentation of Data

### Dataset : cars_data.csv

#### Operations done on ‘cars_data’ dataset: 

1. Read the dataset to the python environment. 

2. Check for the null values present in the dataset. 

3. Plot a bar graph of male vs female buyers participated in the sales. 

4. Find the top 5 cars based on their sales price. 

5. Find the least 5 cars based on their Resell price. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Reading the Dataset

```python
data_df=pd.read_csv("cars_data.csv")
data_df.head()
```

## 2.Check for null values present in the dataset

```python
data_df.isnull().sum()
```

## 3. Bar graph Male vs Female buyers

```python
counts = data_df['Buyer Gender'].value_counts().to_frame().reset_index()
counts
```

```python
plt.figure(figsize=(5,6))
plt.bar(counts['index'],counts['Buyer Gender'],color = ['blue','green'])
plt.title('Bar graph - Male vs Female buyers in the sales', fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Frequency') 
plt.ylim(0,6000)
for i,j in zip(counts['index'],counts['Buyer Gender']):
    plt.annotate(str(j),(i,j+50))
plt.show()
```

### Insights:
1. 104 Female buyers are more in sales than Male buyers.

## 4.Top 5 cars based on the Sales Price

```python
data_df.nlargest(5, ['Sale Price'])
```

## 5. Least 5 cars based on the Resell Price

```python
data_df.nsmallest(5,['Resell Price'])
```
