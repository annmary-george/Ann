## Case Study on ANOVA
### XYZ Company has offices in four different zones. The company wishes to investigate the following :
* The mean sales generated by each zone.
* Total sales generated by all the zones for each month.
* Check whether all the zones generate the same amount of sales.
#### Help the company to carry out their study with the help of data provided. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import scipy.stats as stats
```

```python
#Reading the dataset
data_df= pd.read_csv('Sales_data_zone_wise.csv')
data_df
```

### 1. The mean sales generated by each zone.

```python
print('Mean of Zone - A =', data_df['Zone - A'].mean())
print('Mean of Zone - B =', data_df['Zone - B'].mean())
print('Mean of Zone - C =', data_df['Zone - C'].mean())
print('Mean of Zone - D =', data_df['Zone - D'].mean())
```

### Insights
* The highest average sales is generated in Zone - D and the lowest average sales is generated in Zone - A.

### 2.Total sales generated by all the zones for each month.

```python
def sum_by_column(dataframe,  cols, new_col):
    dataframe[new_col] = dataframe[cols].astype(float).sum(axis=1)
    return(dataframe)

sum_by_column(data_df,  ['Zone - A','Zone - B','Zone - C','Zone - D'],'Total_sales')
```

### 3.Check whether all the zones generate the same amount of sales.

#### Hypothesis :-
* Null Hypothesis :- All the zones generate same amount of sales.
* Alternate Hypothesis :- All the zones generate different amount of sales.

```python
stats.f_oneway(data_df['Zone - A'],data_df['Zone - B'],data_df['Zone - C'],data_df['Zone - D'])
```

### Conclusion :
Since the pValue(0.001182) is much lesser than the alpha value(0.05), we can Reject the Null Hypothesis and conclude that all the 4 zones generate different amount of sales.
